```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Weeks_Claimed_in_All_Programs.csv')
```

```python
df.dropna(subset=["Total"])
df['Reflect_Week'] = pd.to_datetime(df['Reflect_Week'])
df['Year'] = df['Reflect_Week'].dt.year
df_18_21=df[df.Year>=2018]
df_18_21.drop('Year',axis='columns', inplace=True)
df_18_21
```

```python
df_18_21["covid_UI"]=df_18_21["PUA"]+df_18_21["PEUC"]
df_18_21
```

```python
plt.rcParams['figure.figsize'] = (10,5)
plt.style.use('fivethirtyeight')
# df_18_21.plot(x='Reflect_Week',y='Total',color='red',linewidth = 2,fontsize = 8)
# df_18_21.plot(x='Reflect_Week',y='covid_UI',color='yellow',linewidth = 2,fontsize = 8)
plt.plot(df_18_21.Reflect_Week,df_18_21.Total,label = "All Programs")
plt.plot(df_18_21.Reflect_Week,df_18_21.covid_UI,label = "Covid UI")
plt.legend()
plt.show()
```
