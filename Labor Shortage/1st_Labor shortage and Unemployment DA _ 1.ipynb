```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('dataset_percentage.csv')
df2 = pd.read_csv('dataset_thousands.csv')
```

```python
newdf1 = df1.drop('Labor Force Participation Rate', 1)
df2.head()
```

```python
plt.rcParams['figure.figsize'] = (10,5)
plt.style.use('fivethirtyeight')
ax = newdf1.plot(linewidth = 2, fontsize = 12)
ax.set_xlabel('DATE')
ax.legend(fontsize = 12)
```

```python
high_ur = newdf1['Unemployment Rate'].where(lambda a: a==a.max()).dropna()
high_asian = newdf1['Unemployment Rate - Asian'].where(lambda a: a==a.max()).dropna()
high_his = newdf1['Unemployment Rate - Hispanic or Latino'].where(lambda a: a==a.max()).dropna()
high_blk = newdf1['Unemployment Rate - Black or African American'].where(lambda a: a==a.max()).dropna()
high_men = newdf1['Unemployment Rate - Men'].where(lambda a: a==a.max()).dropna()
high_women = newdf1['Unemployment Rate - Women'].where(lambda a: a==a.max()).dropna()
print(high_ur, high_asian, high_his, high_blk, high_men, high_women)
```

```python
print(df1['Labor Force Participation Rate'].where(lambda a: a==a.min()).dropna())
df1['Labor Force Participation Rate'].plot()
```
